**Answering with Agents**

```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager #后面很多函数都没有调用到，在这里先import进来，确保前面autogen库已经有了
```

```python
llm_config = {
    "config_list": [{"model": "gpt-4", "api_key": os.environ["OPENAI_API_KEY"]}],

}
```

```python

Calculation_Financial_Agent = AssistantAgent(
    name="Calculation_Financial_Agent",
    llm_config=llm_config,
    system_message="""
    Strict Rule: Give the correct answer under 5000 tokens

    You are a calculation expert for financial formulas. Show the calculations, explain and give answer.

    """,
)

Financial_Context_Selector = AssistantAgent(
    name="Financial_Context_Selector",
    llm_config=llm_config,
    system_message="""
    Strict Rule: Give the correct sentences under 5000 tokens

    You are expert in identifying the correct sentences for the financial question asked which can used to answer the question. Given the question, identify the relevant sentences .

    """,
)


user_proxy = UserProxyAgent(
    name="User_proxy",
    system_message="A human admin.",
    code_execution_config={
        "last_n_messages": 2,
        "work_dir": "groupchat",
        "use_docker": False,
    },  # Please set use_docker=True if docker is available to run the generated code. Using docker is safer than running the generated code directly.
    human_input_mode="TERMINATE",
)




# define group chat
groupchat = GroupChat(agents=[user_proxy, Financial_Context_Selector, Calculation_Financial_Agent ], messages=[], max_round=4,speaker_selection_method="round_robin")
manager = GroupChatManager(groupchat=groupchat, llm_config=llm_config)
```

```python
def retriever_agent(question: str):
    sources = vectorstore.similarity_search_with_relevance_scores(query, k = 5)
    user_proxy.initiate_chat(
    manager,
    message=f"""Give me the answer to the question: {question}


Here are my full Source Texts closest possible:
{sources}
""")
```

```python
query = "What was Nvidia's net gross profit and loss in 10-k 2016 to 2017?"
retriever_agent(query)
```
